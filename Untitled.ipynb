```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
path = "./experiment_data/"
```

```python
def do_small_plot(datatype):
    path = "./experiment_data/"
    exps = ["SGD_small", "SP2l1_small", "SP2l2_small", "SP2max_small"]

    for exp in exps:
        data_file = path + exp + "/" + datatype + ".pkl"
        data = pickle.load( open(data_file, "rb" ) )
        plt.plot(data, label=exp)

    plt.ylabel(datatype)
    plt.legend()
    plt.show()
```

```python
def do_med_plot(datatype):
    path = "./experiment_data/"
    exps = ["SGD_med", "SP2l1_med", "SP2l2_med", "SP2max_med"]

    for exp in exps:
        data_file = path + exp + "/" + datatype + ".pkl"
        data = pickle.load( open(data_file, "rb" ) )
        plt.plot(data, label=exp)

    plt.ylabel(datatype)
    plt.legend()
    plt.show()
```
